```python
X = data.drop(['Class'], axis=1)    #we are separating the attributes and Class

y = data['Class']                   #assigning Class column to a new variable y
```

```python
from sklearn.model_selection import train_test_split                                   #splitting the training and test datasets
X_train, X_test, y_train ,y_test = train_test_split(X,y,train_size=0.8,random_state=0)
X_train.shape ,X_test.shape
```

```python
cols = X_train.columns
from sklearn.preprocessing import StandardScaler       #scaling the data in order to prevent errors caused by uneven sizing of datas in the knn algorithm

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])
X_train.head()
```

The following piece of code is a knn implementation with two different distance calculations and for different values of k (3,5,7,9)

```python
#the knn implementation
from collections import Counter

def euclidean_distance (point1,point2):
  return np.sqrt(np.sum((np.array(point1)-np.array(point2))**2))     #a function for calculating Euclidean Distance

def manhattan_distance (point1,point2):
  return np.sum(np.abs(np.array(point1)-np.array(point2)))          #a function for calculating Manhattan Distance

def knn_predict_euclidean(training_data,training_labels,test_data,k):    #a function that makes the prediction with knn algorithm by using Euclidean Distance
  distances=[]
  for i in range(len(training_data)):
    dist = euclidean_distance(test_data,training_data[i])             #calling the euclidean distance function and giving the training data points and the test data point
    distances.append((dist,training_labels[i]))                       #adding the calculated distances with the classes of the training data to the distances list
  distances.sort(key = lambda x: x[0])                                #sorting the distances list by just looking at the calcolated distances
  k_nearest_labels=[]
  for i in range(k):
      label=distances[i][1]                                          #getting the labels of the first k elements
      k_nearest_labels.append(label)                                 #storing them in the list of k_nearest_labels
  counts=Counter(k_nearest_labels)
  most_common_label,_=counts.most_common(1)[0]                       #counting the labels to decide which one is the most common one among them
  return most_common_label                                           #then return that prediction you make


def knn_predict_manhattan(training_data,training_labels,test_data,k): #a function that makes the prediction with knn algorithm by using Manhattan Distance
  distances=[]
  for i in range(len(training_data)):
    dist = manhattan_distance(test_data,training_data[i])
    distances.append((dist,training_labels[i]))
  distances.sort(key = lambda x: x[0])
  k_nearest_labels=[]
  for i in range(k):
      label=distances[i][1]
      k_nearest_labels.append(label)
  counts=Counter(k_nearest_labels)
  most_common_label,_=counts.most_common(1)[0]
  return most_common_label


y_predict_3 = []
k=3
for i in range(len(X_test)):
    prediction = knn_predict_euclidean(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_3.append(prediction)


from sklearn.metrics import accuracy_score

print('Model accuracy score for k=3: {0:0.4f}'. format(accuracy_score(y_test, y_predict_3)))

accuracies_euclidean = []
accuracies_euclidean.append( format(accuracy_score(y_test, y_predict_3)))

#to check whether there's overfitting
y_predict_train = []

for i in range(len(X_train)):
    prediction = knn_predict_euclidean(X_train.values, y_train.values, X_train.iloc[i].values, 3)
    y_predict_train.append(prediction)

print('Model accuracy score for training data: {0:0.4f}'. format(accuracy_score(y_train, y_predict_train)))


#for different k values


y_predict_5 = []
k=5
for i in range(len(X_test)):
    prediction = knn_predict_euclidean(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_5.append(prediction)


print('Model accuracy score for k=5: {0:0.4f}'. format(accuracy_score(y_test, y_predict_5)))

accuracies_euclidean.append( format(accuracy_score(y_test, y_predict_5)))

y_predict_7 = []
k=7
for i in range(len(X_test)):
    prediction = knn_predict_euclidean(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_7.append(prediction)


print('Model accuracy score for k=7: {0:0.4f}'. format(accuracy_score(y_test, y_predict_7)))

accuracies_euclidean.append( format(accuracy_score(y_test, y_predict_7)))

y_predict_9 = []
k=9
for i in range(len(X_test)):
    prediction = knn_predict_euclidean(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_9.append(prediction)


print('Model accuracy score for k=9: {0:0.4f}'. format(accuracy_score(y_test, y_predict_9)))

k_values = list(range(3, 10, 2))
accuracies_euclidean.append( format(accuracy_score(y_test, y_predict_9)))
plt.figure(figsize=(10, 8))
plt.plot(k_values, accuracies_euclidean, marker='o', color='blue', linewidth=2, label='Accuracy')
plt.title('Change in accuracy with respect to k(Euclidean Distance)')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.grid(True, linestyle='--', alpha=0.6)
```

🤔 We can see that the accuracy increases with the increase of k. After k=7 it reaches the maximum value and stays in there

```python
#for manhattan distance
y_predict_3m = []
k=3
for i in range(len(X_test)):
    prediction = knn_predict_manhattan(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_3m.append(prediction)


print('Model accuracy score for k=3 Manhattan distance: {0:0.4f}'. format(accuracy_score(y_test, y_predict_3m)))

accuracies_manhattan = []
accuracies_manhattan.append( format(accuracy_score(y_test, y_predict_3m)))


y_predict_5m = []
k=5
for i in range(len(X_test)):
    prediction = knn_predict_manhattan(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_5m.append(prediction)



print('Model accuracy score for k=5 Manhattan distance: {0:0.4f}'. format(accuracy_score(y_test, y_predict_5m)))

accuracies_manhattan.append( format(accuracy_score(y_test, y_predict_5m)))

y_predict_7m = []
k=7
for i in range(len(X_test)):
    prediction = knn_predict_manhattan(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_7m.append(prediction)



print('Model accuracy score for k=7 Manhattan distance: {0:0.4f}'. format(accuracy_score(y_test, y_predict_7m)))

accuracies_manhattan.append( format(accuracy_score(y_test, y_predict_7m)))

y_predict_9m = []
k=9
for i in range(len(X_test)):
    prediction = knn_predict_manhattan(X_train.values, y_train.values, X_test.iloc[i].values, k)
    y_predict_9m.append(prediction)



print('Model accuracy score for k=9 Manhattan distance: {0:0.4f}'. format(accuracy_score(y_test, y_predict_9m)))


k_values = list(range(3, 10, 2))
accuracies_manhattan.append( format(accuracy_score(y_test, y_predict_9m)))
plt.figure(figsize=(10, 8))
plt.plot(k_values, accuracies_manhattan, marker='o', color='blue', linewidth=2, label='Accuracy')
plt.title('Change in accuracy with respect to k (Manhattan Distance)')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.grid(True, linestyle='--', alpha=0.6)
```
